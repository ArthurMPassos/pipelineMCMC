```python
time = time_phased
flux = smoothed_LC

flux_err = numpy.var(flux)
data = (time, flux, flux_err)
nwalkers = 25
niter = 25
initial = numpy.array([u1_p0, u2_p0, a_p0, inc_p0, rp_p0])
ndim = len(initial)
p0 = [numpy.array(initial) + 1e-4 * numpy.random.randn(ndim) for i in range(nwalkers)]

print(p0)
```

```python
#======================================== EXCLUIR =======================================

#--------------------------------------------------#
#----------------------MCMC------------------------#
#--------------------------------------------------#
'''
todo:
 - Arrumar semiEixoUA; V
 - Tirar animação; V
 - Padronizar nome de variáveis; V
 - Chamar ajuste como objeto de classe;
 - Implementar essas funções
'''

def eclipse_mcmc(time, theta):
    rsun = 1.      # TEMPORÁRIO
    u1, u2, semiEixoUA, anguloInclinacao, raioPlanJup = theta
    periodo = 1.
    raioStar, raioPlanetaRstar, semiEixoRaioStar = converte(rsun,raioPlanJup,semiEixoUA)
    
    estrela_ = estrela(373, raioStar, 240., u1, u2, 856)
    Nx = estrela_.getNx()
    Ny = estrela_.getNy()
    raioEstrelaPixel = estrela_.getRaioStar()
    estrelaManchada = estrela_.getEstrela()
    
    eclipse = Eclipse(Nx,Ny,raioEstrelaPixel,estrelaManchada)
    eclipse.setTempoHoras(1.)
    eclipse.criarEclipse(semiEixoRaioStar, semiEixoUA, raioPlanetaRstar, raioPlanJup, periodo, anguloInclinacao, 0,0 , 0, False, False)
    lc0 = numpy.array(eclipse.getCurvaLuz()) 
    ts0 = numpy.array(eclipse.getTempoHoras()) 
    return interpolate.interp1d(ts0,lc0,fill_value="extrapolate")(time)
#--------------------------------------------------#
def lnlike(theta, time, flux, flux_err):
    return -0.5 * numpy.sum(((flux - eclipse_mcmc(time, theta))/flux_err) ** 2)
#--------------------------------------------------#
def lnprior(theta):
    u1, u2, semiEixoUA, anguloInclinacao, rp = theta
    if 0.0 < u1 < 1.0 and 0.0 < u2 < 1.0 and 0.001 < semiEixoUA < 1 and 80. < anguloInclinacao < 90 and 0.1 < rp < 10:
        return 0.0
    return -numpy.inf
#--------------------------------------------------#
def lnprob(theta, time, flux, flux_err):
    lp = lnprior(theta)
    if not numpy.isfinite(lp):
        return -numpy.inf
    return lp + lnlike(theta, time, flux, flux_err)
#--------------------------------------------------#
def main(p0, nwalkers, niter, ndim, lnprob, data):
    sampler = emcee.EnsembleSampler(nwalkers, ndim, lnprob, args=data)

    print("Running burn-in...")
    p0, _, _ = sampler.run_mcmc(p0, 5, progress=True)
    sampler.reset()

    print("Running production...")    
    pos, prob, state = sampler.run_mcmc(p0, niter, progress=True)

    return sampler, pos, prob, state
#--------------------------------------------------#    
def sample_walkers(nsamples,flattened_chain):
    models = []
    draw = numpy.floor(numpy.random.uniform(0,len(flattened_chain),size=nsamples)).astype(int)
    thetas = flattened_chain[draw]
    for i in thetas:
        mod = eclipse_mcmc(time, i)
        models.append(mod)
    spread = numpy.std(models,axis=0)
    med_model = numpy.median(models,axis=0)
    return med_model,spread
#--------------------------------------------------#

#======================================== EXCLUIR ========================================
```

```python
sampler, pos, prob, state = main(p0,nwalkers,niter,ndim,lnprob,data)
```

```python
samples = sampler.flatchain
samples[numpy.argmax(sampler.flatlnprobability)]
```
